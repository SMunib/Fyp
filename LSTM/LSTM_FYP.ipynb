```python
import pandas as pd
items_path = 'items.csv'
transactions_path = 'transactions_2015.csv'
holidays_path = 'holidays_2015.csv'
stores_path = 'quito_stores.csv'
train_path = 'train_subset_2015.csv'
oil_path='oil_prices_filled.csv'

items = pd.read_csv(items_path, dtype={'item_nbr': 'int32', 'family': 'category', 'perishable': 'int8'})
transactions = pd.read_csv(transactions_path, dtype={'store_nbr': 'int16', 'transactions': 'int32'})
holidays = pd.read_csv(holidays_path, dtype={'type': 'category', 'transferred': 'bool'})
stores = pd.read_csv(stores_path, dtype={'store_nbr': 'int16', 'city': 'category', 'state': 'category', 'cluster': 'int8'})
weather=pd.read_csv('weather_2015_filled.csv')
oil=pd.read_csv(oil_path)
# Add a 'holiday' column to holidays DataFrame
holidays['holiday'] = 1  # Mark all rows as holidays

# Read the first 1,000,000 rows from train.csv
train = pd.read_csv(
    train_path,
    dtype={'id': 'int32', 'item_nbr': 'int32', 'store_nbr': 'int16', 'unit_sales': 'float32'},
    nrows=1000000,
)
```

```python


# Load auxiliary files with memory-efficient data types

# Merge with holidays
train = pd.merge(train, holidays, on='date', how='left')
train['holiday'] = train['holiday'].fillna(0).astype(int)

# Merge with items
train = pd.merge(train, items, on='item_nbr', how='left')

# Merge with stores
train = pd.merge(train, stores, on='store_nbr', how='left')

# Merge with transactions
train = pd.merge(train, transactions, on=['store_nbr', 'date'], how='left')
train = pd.merge(train, weather, on='date', how='left')
train = pd.merge(train, oil, on='date', how='left')
# # Fill missing transaction values with 0
train['transactions'] = train['transactions'].fillna(0).astype(int)

# Save the merged DataFrame to a new CSV file
# train.to_csv("merged_data_1M.csv", index=False)
```

```python
train.drop(columns=['type_x', 'locale','locale_name','description','transferred','state','type_y','cluster',], inplace=True)
train.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
data = train.copy()
data.head()
```

```python
# Gives a count of NaNs for each column
data['dcoilwtico'] = data['dcoilwtico'].fillna(data['dcoilwtico'].mean())
print(data.isna().sum())
```

```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
# Extract day of week and week of year
data['day_of_week'] = data['date'].dt.dayofweek
data['week_of_year'] = data['date'].dt.isocalendar().week

# Encode categorical variables
label_encoders = {}
for col in ['family', 'city']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Scale numeric features
scaler = MinMaxScaler()
numeric_cols = ['unit_sales', 'transactions', 'Avg Temperature (C)', 'Min Temperature (C)',
                'Max Temperature (C)', 'Precipitation (mm)', 'perishable', 'day_of_week', 'week_of_year','dcoilwtico']
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
# Define the sequence length
seq_length = 14  # Weekly patterns

# Group data by store and item to create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    grouped_data = data.groupby(['store_nbr', 'item_nbr'])
    for _, group in grouped_data:
        group = group.sort_values(by='date')  # Ensure chronological order
        group = group.reset_index(drop=True)
        values = group[['unit_sales', 'transactions', 'Avg Temperature (C)',
                        'Min Temperature (C)', 'Max Temperature (C)', 'Precipitation (mm)',
                        'perishable','day_of_week', 'week_of_year','dcoilwtico']].values
        for i in range(len(values) - seq_length):
            X.append(values[i:i + seq_length])  # Input sequence
            y.append(values[i + seq_length, 0])  # Target: unit_sales
    return np.array(X), np.array(y)

X, y = create_sequences(data, seq_length)
print(f"Shape of X: {X.shape}")  # (samples, seq_length, features)
print(f"Shape of y: {y.shape}")  # (samples,)
```

```python
print(X[0][0])
```

```python
# Use 80% of data for training and 20% for testing
split_index = int(0.8 * len(X))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"Training samples: {len(X_train)}, Testing samples: {len(X_test)}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Define the LSTM model
model = Sequential([
    LSTM(64, activation='relu', input_shape=(seq_length, X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(64, activation='relu'),
    Dropout(0.2),
    Dense(1)  # Output layer: predict unit_sales
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
# Evaluate the model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test Loss (MSE): {test_loss}, Test MAE: {test_mae}")
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(y_test, y_pred)
accuracy = 100 - mape * 100  # Accuracy as a percentage
print(f"MAPE: {mape:.4f}, Accuracy: {accuracy:.2f}%")
```
